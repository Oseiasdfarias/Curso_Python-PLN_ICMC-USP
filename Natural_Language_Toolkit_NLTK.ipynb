```python
from nltk import bigrams
from nltk import trigrams
from nltk import ngrams
```

### **Bigrams**

```python
tokens = nltk.word_tokenize(corpus)
bigrans_tokens = list(bigrams(tokens))
bigrans_tokens[:20]
```

### **Trigrams**

```python
trigrams_tokens = list(trigrams(tokens))
trigrams_tokens[:20]
```

### **N-grams**

```python
ngrams_tokens = list(ngrams(tokens, 4))
ngrams_tokens[:20]
```

### **Utilização: Reconhecendo Entidades Nomeadas usando os n-grams**

#### **Bigrams**

```python
for bgrams in bigrans_tokens:
  if bgrams[0][0].isupper() and bgrams[1][0].isupper():
    print(bgrams)
```

#### **Bigrams**

```python
for tgrams in trigrams_tokens:
  if tgrams[0][0].isupper() and tgrams[1][0].isupper() and tgrams[2][0].isupper():
    print(tgrams)
```

## **Stemmers e Lematização NLTK**



### **Stemmers**
#### **RSLP: Removedor de Sufixos da Língua Portuguesa**

```python
stemmer = nltk.RSLPStemmer()

lista_st = ["Amigão", "amigo", "amigos", "propuseram", "propõem", "propondo"]
lista_stemmer = []
for i in lista_st:
  lista_stemmer.append(stemmer.stem(i))

print(lista_stemmer)
```

## **ETIQUETADOR**

```python
from nltk.corpus import mac_morpho
from nltk.tag import UnigramTagger

tokens = nltk.word_tokenize(corpus)

sentencas_treino = mac_morpho.tagged_sents()
etiquetador = UnigramTagger(sentencas_treino)

etiquetado = etiquetador.tag(tokens)
```

```python
etiquetado[:20]
```

```python
from nltk.corpus import mac_morpho
from nltk.tag import UnigramTagger
from nltk.tag import DefaultTagger

tokens = nltk.word_tokenize(corpus)

# Dessa vez utilizaremos o DefaultTagger para definir uma etiqueta padrão
etiq_padrao = DefaultTagger('N')
sentencas_treino = mac_morpho.tagged_sents()
etiquetador = UnigramTagger(sentencas_treino, backoff=etiq_padrao)

etiquetado = etiquetador.tag(tokens)
```

```python
etiquetado[:20]
```

### **Buscando padões**

```python
from nltk.chunk import RegexpParser

pattern = 'NP: {<NPROP><NPROP> | <N><N>}'
analise_gramatical = RegexpParser(pattern)

arvore = analise_gramatical.parse(etiquetado)
print(arvore)
```
